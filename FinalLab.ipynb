# Problema 1

```python
ket_plus = Statevector.from_label('+')
ket_plus.draw('Latex')
```

```python
ket_minus = Statevector.from_label('-')
ket_minus.draw('Latex')
```

```python
ket_plus2 = Statevector.from_label('+')
ket_plus2.draw('Latex')
```

```python
ket_system1 = Statevector(ket_plus.tensor(ket_plus2))
Statevector(ket_system1).draw('Latex')
```

```python
ket_system = Statevector(ket_system1.tensor(ket_minus))
Statevector(ket_system).draw('Latex')
```

```python
probabilities = ket_system.probabilities()
print(probabilities)
```

## Intento de poner la compuerta CU3 en forma matricial

```python
ket0 = [[1],[0]]
array_to_latex(ket0)
```

```python
ket1 = [[0],[1]]
array_to_latex(ket1)
```

```python
Ket_0 = Statevector(ket0)
Ket_0.draw('Latex')
```

```python
Ket_1 = Statevector(ket1)
Ket_1.draw('Latex')
```

```python
Operador_ket_0 = Operator(ket0)
array_to_latex(Operador_ket_0)
```

```python
Operador_ket_1 = Operator(ket1)
array_to_latex(Operador_ket_1)
```

```python
Bra0 = [1,0]
array_to_latex(Bra0)
```

```python
Operador_Bra_0 = Operator(Bra0)
array_to_latex(Operador_Bra_0)
```

```python
Bra1 = [0,1]
array_to_latex(Bra1)
```

```python
Operador_Bra_1 = Operator(Bra1)
array_to_latex(Operador_Bra_1)
```

```python
Producto_externo_00 = np.outer(Operador_Bra_0, Operador_ket_0)
Producto_externo_00
```

```python
Producto_externo_11 = np.outer(Operador_Bra_1, Operador_ket_1)
array_to_latex(Producto_externo_11)
```

```python
Identity = Producto_externo_00 + Producto_externo_11
array_to_latex(Identity)
```
